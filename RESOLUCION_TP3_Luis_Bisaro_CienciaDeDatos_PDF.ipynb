## ***EJERCICIO 2***  ✅

```python
########################### EJERCICIO 2 ############################################

import matplotlib.pyplot as plt
import random
import numpy as np

dado = [1,2,3,4,5,6] # defino caras del dado
tiros=[] #guardo los valores de las tiradas
decimales=4 # defino la cantidad de decimales a usar

print('ingrese cantidad de tiros >30') #indico la cantidad de tiros a realizar
cant_tiros=int(input())

for i in range(cant_tiros):
  tiro=random.choices(dado,weights=[2,2,1,1,1,1])
  tiros.append((tiro))
#print(tiros)

valores,cuenta=np.unique(tiros,return_counts=True)
#print('valores=', valores) #guarda los valores unicos en la lista
#print('cuenta=', cuenta)   #guarda la cuenta de cada uno en el mismo orden

uno=round(cuenta[0]/cant_tiros,decimales)
dos=round(cuenta[1]/cant_tiros,decimales)
tres=round(cuenta[2]/cant_tiros,decimales)
cuatro=round(cuenta[3]/cant_tiros,decimales)
cinco=round(cuenta[4]/cant_tiros,decimales)
seis=round(cuenta[5]/cant_tiros,decimales)

print()
print (f"la probabilidad de obtener en {cant_tiros} tiros el valor 2 es P(2)={dos} ,la de obtener 3 es P(3)={tres} ,y la de 5 P(5)={cinco}" )

print()
print()

tirada=np.array([uno,dos,tres,cuatro,cinco,seis])

x= np.array(range(6))
fig, ax = plt.subplots()
plt.bar(x+1, tirada,color='#C0C0C0')

plt.axhline(y=(0.250), color="r", label="probabilidad para 1 y 2 = 0,250")
plt.legend(loc = 'lower right')

plt.axhline(y=(0.125), color="g", label="probabilidad para 3,4,5 y 6 = 0,125")
plt.legend(loc = 'lower right')

for i, total in enumerate(tirada):
  ax.text(i+1, tirada[i]-(0.2*tirada[i]) , tirada[i],
          ha = 'center', weight = 'bold', color = '#FFD700')


plt.ylabel('Frecuencia relativa')
plt.title('Histograma para '+str(cant_tiros)+' tiros')

plt.show()
```

## EJERCICIO 3  ✅

```python
########################### EJERCICIO 3 ############################################

import matplotlib.pyplot as plt
import random
import numpy as np

opcion = (['piedra', 'papel', 'tijera'])# defino opciones
tiros=[] #guardo los valores de las tiradas
decimales=4 # defino la cantidad de decimales a usar

print('ingrese cantidad de tiros') #indico la cantidad de tiros a realizar
cant_tiros=int(input())

for i in range(cant_tiros):
  tiro=random.choices(opcion,weights=[1,1,1])
  tiros.append((tiro))
#print(tiros)

valores,cuenta=np.unique(tiros,return_counts=True)
#print('valores=', valores) #guarda los valores unicos en la lista
#print('cuenta=', cuenta)   #guarda la cuenta de cada uno en el mismo orden

piedra=round(cuenta[0]/cant_tiros,decimales)
papel=round(cuenta[1]/cant_tiros,decimales)
tijera=round(cuenta[2]/cant_tiros,decimales)

print()
print (f"la probabilidad de obtener en {cant_tiros} tiros, piedra es P()={piedra} ,la de obtener papel es P()={papel} ,y la de tijera P()={tijera}" )
print()
print()

tirada=np.array([piedra,papel,tijera])

x= np.array(range(3))
fig, ax = plt.subplots()
plt.bar(opcion, tirada,color = 'grey')

plt.axhline(y=(0.333333), color="r", label="probabilidad para cada opcion es 1/3")
plt.legend(loc = 'lower right')

plt.ylabel('Frecuencia relativa')
plt.title('Histograma para '+str(cant_tiros)+' tiros')

for i, total in enumerate(tirada):
  ax.text(i, tirada[i]-(0.2*tirada[i]) , tirada[i],
          ha = 'center', weight = 'bold', color = 'black')


plt.show()


```
