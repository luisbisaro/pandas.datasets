###CODIGO B.2: Creacion de un perceptron de formal manual

```python
###################################################################################
#______________ B.2 _________________________
###############################################
#_______________PROCESO DE CREACION DE PERCEPTRON- de forma manual con codigo___________________

#IMPORTANTE:
#Casi el mismo PROCESO que uds aplican para hacer un MODELO DE PREDICCION (Regresion, clasificacion, etc)
#es APROXIMADAMENTE, el mismo proceso que se realiza cuando CREAMOS, ACTIVAMOS Y ENTRENAMOS UNA NEURONA


##############################################
#Importar LIBRERIAS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


################################################################################################################################################
#__________________________#Aqui van LAS FUNCIONES HECHAS POR NOSOTROS. (Creamos funciones) #Aqui crearemos la neurona_____________________

#=============================================
#¿Que es una funcion de ACTIVACION?:

      #UNa funcion de activacion () DECIDE si una NEURONA ARTIFICIAL debe ACTIVARSE (enviar una señal o no)
          #en funcion de su entrada ponderada


      #Es un INTERRUPTOR (Algorimo matematico)


#=============================================
#Fucion de ACTIVACION:
def activacion(valor):
      if valor>= 0:
          return 1
      else:
          return 0



#=============================================
#Funcion para ENTRENAR al PERCEPTRON
def entrenar_perceptron(X, Y, lr=0.1, epochs=1000): #CANTIDAD DE ITERACIONES, CESAR Nuñez: "El numero de veces que se van a pasar cada ejemplo de entrenamiento por la red",
                                                    #Cantidad de veces que se pasa por la caja de entrenamiento, regulando los pesos (prioridades e importancias que le enseñamos al perceptron= neurona basica)
      num_entradas= X.shape[1] #¿?
      #print("\nBANDERA: num de entradas")
      #print(num_entradas)

      num_muestras=X.shape[0]

      #Pesos conceptualmente a modo gauchesco es= LA IMPORTANCIA, a donde queremos que filtre o pese mas. Patron u objetivo
      pesos= np.random.rand(num_entradas) #Creo numeros al azar

      for epoch in range(epochs):
          for i in range(num_muestras):
                entrada= X[i]
                salida_real= Y[i]

                salida_predicha= activacion(np.dot(entrada, pesos)) #Esta funcion INTERNA recibe (ENTRADA, PESO) y se convierte en un numero, el cual es enviado a LA FUNCION DE ACTIVACION
                error= salida_real - salida_predicha

                pesos= pesos+ (lr*error*entrada)

      return pesos




##############################################################################################################################################
#_______________________#Aqui EN LUGAR en el que VAMOS A INVOCAR LAS FUNCIONES #Aqui inicia el codigo____________________

########################
#==============================================================
#NOTA
#Crear UNA NEURONA,
# Es el mismo proceso que uds APLICAN cuando APLICAN ALGUN MODELO DE PREDICCION (regresion, clasificacion etc)
#LA DIFERENCIA esta en que en vez de APLICAR EL MODELO DE PREDICCION, aplican LA NEURONA
#==============================================================


##############################################
#Cargar el DATA SET de DIGITOS:
digits= datasets.load_digits()


############################################################
#Exploracion del DATA SET (1 lo exploras como data set, 2 Lo convertis en un DF y lo exploras):

#....Escribir aqui su codigo para probar......


#############################################################
#Limpieza (Como estamos trabajando con un data set de la lIBRERIA SKERLEARNING)
#Este data set ya fue limpiado cuando lo incluyeron

#......No habria que hacer limpieza.........


##############################################
#Eligo mi X y mi Y:
X= digits.data #Investigar el data set de digits y saber que es este "data"
Y= digits.target #Investigar el data set de digits y saber que es este "target"

#print("\nBANDERA: X")
#print(X)

#print("\nBANDERA: X")
#print(Y)

##############################################
#Divido el CONJUNTO de DATOS en (ENTRENAMIENTO Y TESTEO)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


##############################################
#Entreno el PERCEPTRON
pesos_entrenados= entrenar_perceptron(X_train, Y_train) #ENvio a la FUNCION DE ENTRENAMIENTO mis TRAIN.


##############################################
#Aplico la NEURONA al CONJUNTO DE PRUEBA:
Y_pred= [activacion(np.dot(x, pesos_entrenados)) for x in X_test] #0



##############################################
#Calculo la PRECISION DEL MODELO:
exactitud= accuracy_score(Y_test, Y_pred)

print("Exactitud: ",exactitud)



######################################################
#==============================================================
#IMPORTANTE:
#PARA ENTENDER ESTE PUNTO, necesitan aprender bien, como esta constituido EL DATA SET DE DIGITOS (PIxel)
#==============================================================


#EXPLORA QUE ESTAMOS POR MOSTRAR:


#Realizo PRUEBAS
#Visualizando algunas PREDICCIONES:

print("\#################################")
print("\nBANDERA FINAL")

print("\n",X_test)
print("\n",X_test[4])

######################################################
#MOSTRARMOS ¿que mostramos?: ejecutar el bloque de abajo







```

```python
######################################################
#Mostramos EL RESULTADOS de 1 una PREDICCION que realiza el PERCEPTRON:

plt.figure(figsize=(10, 4))

plt.subplot(2, 5, 4)
plt.imshow(X_test[4].reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')
plt.title(f"prediccion: {Y_pred[4]}")
plt.show()

```

```python
######################################################
#MOSTRARMOS LOS RESULTADOS de la PREDICCION CON UN CICLO de las prediccion que realiza el PERCEPTRON= RED NEURONAL:

plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_test[i].reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'Predicción: {Y_pred[i]}')
plt.show()
```

##ETAPA 2: SEGUNDA FORMA DE CREACION DE PERCEPTRON- con tensorflow (Visualizamos LA CANTIDAD DE VECES QUE AJUSTA EL MODELO= EPOCH = BACKPROPAGATION)

```python
###############################################
#_______________PROCESO SEGUNDA FORMA DE CREACION DE PERCEPTRON- con tensorflow___________________

#IMPORTANTE:
#Casi el mismo PROCESO que uds aplican para hacer un MODELO DE PREDICCION (Regresion, clasificacion, etc)
#es APROXIMADAMENTE, el mismo proceso que se realiza cuando CREAMOS, ACTIVAMOS Y ENTRENAMOS UNA NEURONA



###################################################################
#Importamos LIBRERIAS
import numpy as np
import tensorflow as tf
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


###################################################################

###################################################################
#Cargar el DATA SET de DIGITOS:
digits= datasets.load_digits()


###################################################################

#Exploracion del DATA SET (1 lo exploras como data set, 2 Lo convertis en un DF y lo exploras):

#....Escribir aqui su codigo para probar......



###################################################################
#Limpieza (Como estamos trabajando con un data set de la lIBRERIA SKERLEARNING)
#Este data set ya fue limpiado cuando lo incluyeron

#......No habria que hacer limpieza.........



###################################################################
#Eligo mi X y mi Y:
X= digits.data #Investigar el data set de digits y saber que es este "data"
Y= digits.target #Investigar el data set de digits y saber que es este "target"



###################################################################
#Divido el CONJUNTO de DATOS en (ENTRENAMIENTO Y TESTEO)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)



###################################################################
#Creamos un PERCEPTRON SIMPLE en tensorflow
model= tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(X_train.shape[1],)))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))


###################################################################
#Compilamos el  MODELO:
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])



###################################################################
#Entrenamos el MODELO con backpropagation:
 #(Si este fuera la creacion de UN PERCEPTRON de formal manual)
 #cuando apliquemos BACKPROPAGATION MANUALMENTE, deberiamos codificar a mano este proceso:

      #(Backpropagation= Crear, activar, entrenar NEURONA, aplicar NEURONA, ver resultados, REMODELAR LA NEURONA(Recodificar), activar, entrenar NEURONA, aplicar NEURONA,
      #ver resultados, REMODELAR NEURONA(Recodificar) y asi sucesivamente.....)

  #Pero como esto ya es una FUNCION en TENSORFLOW, directamente aplicamos la FUNCION BACKPROPAGATION y listo.
model.fit(X_train, Y_train, epochs=100, verbose=1)


###################################################################
#Realizamos las PREDICCIONES en el conjunto de prueba
Y_pred= (model.predict(X_test)>0.5).astype(int)



###################################################################
#Calcular la PRECISION del MODELO:
#Estamos calculando LA PRESICION que estan REALIZANDO LAS ITERACIONES = EPOCHS

#Bere= "Indica la cantidad de veces que realiza backpropagation"

exactitud= accuracy_score(Y_test, Y_pred)

print("\nExactitud: ",exactitud)

#=================================================




###################################################################
#TAREA: Visualizar (Copiando o inspirandose del codigo de LA ETAPA 1 o del
#CODIGO A) lo que el MODELO PREDICE, es decir DIBUJARLO, PLOTEARLO







```

##ETAPA 3: Ejemplo DEMOSTRATIO de que haga PREDICCION con un DIBUJO QUE LE ENTREGUEMOS (El dibujo de codigo es manual esta hecho en una matriz y dibuja numeros)
Creamos un PERCEPTRON simple, y trataremos de introducir una MATRIZ (8X8), de un digito para saber si el MODELO LO PODRIA PREDECIR.


```python
###################################################################
#_______________________________Creamos un PERCEPTRON para que entienda un DIBUJO HECHO POR NOSOTROS_____________________
#El dibujo esta hecho en una matriz de codigo, y dibujamos numero


#==================
#PROCESO (Inspirarse de AQUI PARA EL INTEGRADOR 2):

    #1- Ingreso el data set
    #2- Elijo mi X e Y
    #3- Divido el CONJUNTO (Train y Test)
    #4- Creamos el PERCEPTRON=MODELO
    #5- Compilo el MODELO=PERCEPTRON
    #6- Entreno el PERCEPTRON=MODELO

    #7- Le doy un nuevo DATO (dibujo en matriz con forma de numero)
    #8- el PERCEPTRON PROCESA el dato nuevo
    #9- Te PREDICE dentro del proceso, lo que quisiste dibujar, y te da una GRAFICA de lo que APROXIMA(predice) QUISISTE DIBUJAR.  (IDENTIFICA que es un 9)

#==================
###################################################################
###################################################################
###################################################################
#Importamos LIBRERIAS
import numpy as np
import tensorflow as tf
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split

###################################################################
###################################################################
###################################################################
#Cargamos el CONJUNTO DE DATOS DE DIGITOS
from sklearn import datasets
digits = datasets.load_digits()


###################################################################
###################################################################
###################################################################
#Exploracion del DATA SET (1 lo exploras como data set, 2 Lo convertis en un DF y lo exploras):

#....Escribir aqui su codigo para probar......


###################################################################
###################################################################
###################################################################
#Limpieza (Como estamos trabajando con un data set de la lIBRERIA SKERLEARNING)
#Este data set ya fue limpiado cuando lo incluyeron

#......No habria que hacer limpieza.........


###################################################################
###################################################################
###################################################################
#Eligo mi X y mi Y:
X= digits.data #Investigar el data set de digits y saber que es este "data"
Y= digits.target #Investigar el data set de digits y saber que es este "target"


###################################################################
###################################################################
###################################################################
#Divido el CONJUNTO de DATOS en (ENTRENAMIENTO Y TESTEO)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)



###################################################################
###################################################################
###################################################################
#Creamos un PERCEPTRON SIMPLE en tensorflow
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(X_train.shape[1],)))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

###################################################################
###################################################################
###################################################################
#Compilo el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


###################################################################
###################################################################
###################################################################
#Entreno el MODELO CON BACKPROPAGATION (Mirar la ETAPA 1, alli estan las entrañas del perceptron, BACKPROPAGATION= Epoch= Reajuste)
model.fit(X_train, Y_train, epochs=50, verbose=1)


#=============================================
#ITRODUCIREMOS nuestro DIBUJO:
#Lo que hicimos anteriormente fue CREAR y ENTRENAR una NEURONA con el DATA SET DE DIGITOS(Datos anteriores)
#Lo que haremos A CONTINUACION es, ya que la neurona sabe aproximadamente como GRAFICAR un 1, 2,3,4, etc (a partir de datos del data set de digitos,
#el cual recordemos que esta muy INEXACTO e imperfecto, osea crudo)

#Ahora LA PONDREMOS A PRUEBA y le daremos UN DIBUJO(este dibujo no es mas que una matriz de numeros que asemeja la forma de un numero)
# y como ya esta entrenada en SABER APROXIMADAMENTE como es EL GRAFICO DE UN DIBUJO de NUMEROS
#(NOTA: Recordar que los numeros que se grafican SON EN REALIDAD MATRICEZ FILAS y COLUMNAS de:
  # (numeros, Coordenadas, Numeros que indican pixeles= columnas y filas, como quieran llamarle pero en si son NUMEROS.)

#podra ENTENDER(PREDECIR) nuestro dibujo y GRAFICAR(mostrar en MATRIZ) lo que quisimos dibujar
 #(aqui es en donde la neurona termina DIBUJANDO(predecir(es decir predice a partir del data set que se le entrego)) MEJOR QUE NOSOTROS..jejeje)

number_9 = [
            [0, 0, 0, 16, 16, 16, 0, 0],
            [0, 0, 16, 0, 0, 16, 0, 0],
            [0, 16, 0, 0, 0, 16, 0, 0], #numero 9
            [0, 0, 16, 16, 16, 16, 0, 0],
            [0, 0, 0, 0, 0, 16, 0, 0],
            [0, 0, 0, 0, 0, 16, 0, 0],
            [0, 0, 0, 0, 0, 16, 0, 0],
            [0, 0, 0, 16, 16, 0, 0, 0]
]

print("\n###################3")
print("\nEl dibujo que hize de un 9 con una MATRIZ de numeros")
print(number_9)
print("\n###################3")


#=============================================
#Redimensiono en DATOS que se asemejen al data set de digitos (Ya que este es el tipo de dato que viene manejando el perceptron)
mi_dibujo= np.array(number_9).flatten()


###################################################################
###################################################################
###################################################################
#Realizar la PREDICCION
prediction= model.predict(np.array([mi_dibujo])) #model=PERCEPTRON


###################################################################
###################################################################
###################################################################
print("\n\n\n\n##################################################")
print("##################################################")
print("\n_____________________________RESULTADOS_______________________")

print("\n##################################################")
print("##################################################")
print("\nMostrar IMAGEN del NUMERO que esta en el DATA SET de digitos: (El dibujo CRUDO)\n\n")
#nota:Recuerda que el dibujo son NUMEROS.

#Obtener la PRIMERA IMAGEN DE UN DIGITO:
primer_digito= digits.images[9]

#MOSTRAR LO OBTENIDO:
plt.figure(figsize=(4, 4))
plt.imshow(primer_digito, cmap=plt.cm.gray_r, interpolation='nearest')
plt.title(f"Ejemplo de digito: {digits.target[9]}")
plt.show()



###################################################################
###################################################################
###################################################################
print("\n\n\n\n##################################################")
print("##################################################")
#TAREA: Mostrar FRMES a traves de un CICLO WHILE de como el PERCEPTRON VA DIBUJANDO de a poco el numero:
print("Mostrar FRMES a traves de un CICLO WHILE de como el PERCEPTRON VA DIBUJANDO de a poco el numero:\n\n")

print("\n.....Escribir CODIGO AQUI.......\n")


###################################################################
###################################################################
###################################################################
print("\n\n\n\n##################################################")
print("##################################################")
print("Mostrar NUESTRO DIBUJO hecho con NUMEROS, en una MATRIZ con la forma de un 9: (nuestro dibujo crudo)\n\n")

print(mi_dibujo)
print("\n#=============================================")
print("\nAqui abajo la forma de 9 que dibujamos: \n")
print("\n[0, 0, 0, 16, 16, 16, 0, 0]\n[0, 0, 16, 0, 0, 16, 0, 0]\n[0, 16, 0, 0, 0, 16, 0, 0]\n[0, 0, 16, 16, 16, 16, 0, 0]\n[0, 0, 0, 0, 0, 16, 0, 0]\n[0, 0, 0, 0, 0, 16, 0, 0]\n[0, 0, 0, 0, 0, 16, 0, 0]\n[0, 0, 0, 16, 16, 0, 0, 0]")




###################################################################
###################################################################
###################################################################
print("\n\n\n\n##################################################")
print("##################################################")
print("Mostrar RESULTADO FINAL, plotear los DATOS que el PERCEPTRON ESTA PREDICIEND (es decir los RESULTADOS DE ESTA PRUEBA) :")
#Mostrar IMAGEN(cordenadas predecidas) PREDECIDA
plt.imshow(np.array(mi_dibujo).reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')

###################################################################
###################################################################
###################################################################
#Mostrar la PREDICCION (Que el perceptron a traves de la prediccion nos grafique lo que quisimos dibujar)

if prediction >0.5:
  plt.title("prediction: 1")

else:
  plt.title("Prediction NO ES 1")

plt.show()


```
